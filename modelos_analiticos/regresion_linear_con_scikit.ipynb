```python
import pandas as pd
import matplotlib.pyplot as plt
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
```

```python
dataset = pd.read_csv('./datasets/salarios.csv')
```

```python
dataset.head()
```

```python
dataset.shape
```

```python
x = dataset.iloc[:, :-1].values #tomo la ultima columna
y = dataset.iloc[:, 1].values #tomo la primera columna
```

```python
#dividimos parte de pruebas y parte de entrenamiento
X_train, X_test, Y_train, Y_test = train_test_split(x,y, test_size = 0.2, random_state=0)
```

```python
#inicializo el algoritmo
regressor = LinearRegression()
```

```python
#entreno el algoritmo con las variables de entrenamiento
regressor.fit(X_train, Y_train)
```

```python
#Usando el set de entrenamiento, predecimos el valor de salario segun años de experiencia
viz_train = plt
viz_train.scatter(X_train, Y_train, color='blue')
viz_train.plot(X_train, regressor.predict(X_train), color='black')
viz_train.title('Salario vs Exp')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```python
#Usando el set de prueba, predecimos el valor de salario segun años de experiencia
viz_train = plt
viz_train.scatter(X_test, Y_test, color='blue')
viz_train.plot(X_train, regressor.predict(X_train), color='black')
viz_train.title('Salario vs Exp')
viz_train.xlabel('Experiencia')
viz_train.ylabel('Salario')
viz_train.show()
```

```python
#verifico el score del algoritmo , es decir, el porcentaje de aciertos,  dara un valor entre 0 y 1
regressor.score(X_test,Y_test)
```
