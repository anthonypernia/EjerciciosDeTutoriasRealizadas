```python
df_join_stats_salary = pd.merge(df_seasons_stats_2017, df_nba_season1718_salary, on=['Player','Tm'], how='inner')
df_join_stats_salary
```

```python
pd.read_sql_query("""SELECT 
                  A.Player AS PLAYER, 
                  ROUND(B.season17_18) AS PROM_SAL, 
                  ROUND(A.PTS) AS PROM_PTS ,
                  CAST(A.Year AS INT) AS YEAR
                  FROM Seasons_Stats AS A 
                  INNER JOIN NBA_season1718_salary AS B 
                  ON A.Player=B.Player
                  WHERE A.Year==2017
                  GROUP BY 1
                  
                  """, db_conn)
```

```python
df_join_stats_salary['Age'] = df_join_stats_salary['Age'].astype(int)
```

```python
df_join_stats_salary
```

```python
df_summarized_by_player = df_join_stats_salary.groupby('Player').sum()
df_summarized_by_player
```
