```python

dfMerged = dfp.merge(dfw, on='ID', how='inner')
```

```python
dfMerged.dtypes
```

```python
dfMerged['Target'].value_counts()
```

```python
def target_solution(target):
    if '<=50k' in str(target).lower():
        return '<=50K'
    else:
        return '>50K'
```

```python
dfMerged['Target'] = dfMerged['Target'].apply(target_solution)
```

```python
dfMerged['Target'].value_counts()
```
