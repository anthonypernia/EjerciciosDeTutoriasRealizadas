```python
import pandas as pd
import sqlalchemy as db
import requests
import json
```

```python
engine = db.create_engine('sqlite:///datasets/nba_salary.sqlite')
connection = engine.connect()
```

```python
dfStats = pd.read_sql_query("SELECT * FROM Seasons_Stats", connection)
dfSalary = pd.read_sql_query("SELECT * FROM NBA_season1718_salary", connection)
```

```python
pd.read_sql_query("SELECT * FROM NBA_season1718_salary WHERE season17_18 > 10000000", connection)
```

```python
pd.read_sql_query("SELECT * FROM Seasons_Stats WHERE Year == 2017", connection)
```

```python
pd.read_sql_query("SELECT Year , ROUND(AST) AS PROM_AST, ROUND(PTS) AS PROM_PTS, ROUND(PF) AS PROM_PF FROM Seasons_Stats GROUP BY 1 ", connection)
```

```python
pd.read_sql_query("""SELECT 
                  A.Player AS PLAYER, 
                  ROUND(B.season17_18) AS PROM_SAL, 
                  ROUND(A.PTS) AS PROM_PTS 
                  FROM Seasons_Stats AS A 
                  INNER JOIN NBA_season1718_salary AS B 
                  ON A.Player=B.Player
                  GROUP BY 1""", connection)
```
